# Basics- Slicing and Filtering

Perhaps the Most common activities performed on dataframes.
using the same data from AirBnB for this example:-https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data

```python
import pandas as pd
import numpy as np

df=pd.read_csv("AB_NYC_2019.csv")
df.head(2)
```

## Slicing columns

```python
# option 1
df["host_name"]
```

```python
# option 2
df.host_name
```

```python
# Multiple Columns

df[['host_name','neighbourhood_group']]
```

## Filtering on rows (mask filtering)

```python
#option 1- in one line
df[df.host_name=="Taz"] #this returns the values in df where 'taz' is the host
```

```python
df.host_name=="Taz"#running just this will give a index and boolean (true,false) for rows where 'taz' is the host name
```

 ##### sum()

```python
(df.host_name=='Taz').sum() # sums up the count of rows conating the value
```

```python
# option 2 using a Mask variable
mask= df.host_name=="Taz"
df[mask].head(2)
```

# -
 Now Lets Use the mask to find Information
 Here we are looking for the Quickest and cheapest rooms
 so filtering the price <1000 and where minimum allowed stay is less that 3 nights
 
 sum-gives us a count of how many such options are availible

```python
quick_n_cheap= (df.price<100) & (df.minimum_nights<3)
quick_n_cheap.sum()
```

```python
df[quick_n_cheap].head(2)
```

Now lets Filter by reviews

```python
reviews_consistent=df[(df.reviews_per_month>3)|(df.number_of_reviews>50)] #3 reviews/month OR total no. of reviews >50
reviews_consistent.head(3)
```

```python
# using Numpy

mask=np.logical_or((df.reviews_per_month>3),(df.number_of_reviews>50))
df[mask].head(2)
```

```python
"""The '~' operator reverses the logic, ie- it will give out all the values that doesnt belong in the specified mask filter"""
df[~mask].head(2) 
```

# Filtering columns and rows together

The method is `.loc`

```python
df.loc[mask,"name"] # This gives all the 'name' satisfying the mask
```

```python
# multiple columns
df.loc[mask,["name","host_name"]].head() # put multiple column names as a list
```

```python
df.loc[mask,:].head(3) # This will give all the columns (for the rows satisfying the mask) as none is specified
```

# Filtering based on index?

```python
"""iloc is based on index, but not on a pandas set index, it follows only the internal array index"""
df.iloc[0,:] 
```

```python
df.iloc[0,1] # gives the first row and the first column
```

```python
df2=df.set_index("id") # ie even if we set a index on a different column it will stil give just the first row
df2.iloc[0,:]
```

```python
df2.iloc[1:4,6:] # this will give you the columns 1 to 4 and the from 6 onwards
```

# Provided mask helpers

```python
df.loc[df.price.isin([100,200]),"price"].head()
```

```python
df.loc[df.price.between(100,200),"price"].head()
```

# Views Vs Copy

A common pitfall for users to not understand the differnce between views and copies.

```python
df2= df.copy() # get a copy of the original df
```

```python
df2['name'][0]="TESTING"
df2.head(1)
```

```python
df2.loc[df2.index==0,"name"]="TESTING2"
df2.head(1)
```

```python
df2[df2.host_name=="Jhon"]["name"]="Jhonny"
df2.head(1)
```
