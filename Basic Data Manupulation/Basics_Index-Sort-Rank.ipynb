# Indexing

    So this means a lot of things depending on the context.
    For Pandas the index is the number of the left, which is the unique value that can identify each row.
    - like the Excel data row numbers 
    By default the index is generated by counting up from zero.
    but in this data we can see the database  index(which is called the primary key)id would also be another good choice.

```python
df2=df.set_index("id") # here we set a index specific to our needs 'id'
df2.head(3)
```

```python
"""here we are using the Index ('id' number) directly to get the value """
df2.name[2539] 
```

```python
"""now we will be using the room type as the index"""
df3=df.groupby("room_type").mean()
df3
```

### Reset Index

```python
"""This will reset the Index to basic count from Zero and returning 'room_type' to just a column header again"""
df3.reset_index() 
```

```python
df3.reset_index(drop=True)
```

# Sorting

Always try and use sort_index for sorting
If i want the df sorted. use sort_values.
sort_index after setting it.

```python
df3.sort_index(ascending=False) #default is ascending=True
```

```python
"""sorting by a specific column"""
df.sort_values("host_name")
df.head(3)   
```

```python
"""multiple Values can be passed to the [sort_values()] function in a list format"""
df.sort_values(["neighbourhood_group","host_name"])# first sorts by "neighbourhood_group" then by "host_name"
df.head(3)
```

```python
"""We can also revese/control the order of sorting for each variable"""
df.sort_values(["neighbourhood_group","host_name"],ascending=[False,True])
df.head(3)
```

However All these chages are temporay and dosent change the original dataFrame

to make chages to the dataframe the argument (`inplace=True`) must be set

eg:-`df.sort_values(["neighbourhood_group","host_name"],ascending=[False,True],inplace=True)`

### unique()

This will give all the unique values present in that column

```python
df.neighbourhood_group.unique()
```

### Value_counts

```python
"""counts the Total Number of entries of each unique values"""
df.neighbourhood_group.value_counts() 
```

# RANK
 
    Like Sorting but with colision detection
    ie:- If 2 or more keys have have same value

```python
"""First lets do a basic sorting to see similar prices, in decending order-[ascending=False]"""
dfp=df.sort_values("price",ascending=False) #This should give us the priciest of rooms
dfp[['id','host_name','price']].head()
```

##### now lets solve this with [Rank]

```python
#df.price.rank(method="average",ascending=False)

"""this will give a new rank column and the prices beside which is not very useful-uncomment to try """
"""So lets try storing in a new column nd then add column beside the values"""
```

```python
 dfp["price_rank"]= df.price.rank(method="average",ascending=False) # storing the ranks in a new column price rank
# dfp["price_rank"]= df.price.rank(method="min",ascending=False)
# dfp["price_rank"]= df.price.rank(method="max",ascending=False)
dfp[['id','host_name','price',"price_rank"]].head()
```
