# Basic Plots

We'll be using the Kaggle Heart Disease UCI dataset as an example. You can find it here:
https://www.kaggle.com/ronitf/heart-disease-uci/


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df=pd.read_csv("heart.csv")
```

# Bar Plots
Comparing discrete data. Availiable through either pandas or matplotlib.

# Pandas

```python
df.head()
```

```python
chest_pain=df.groupby(by="cp").median().reset_index()
chest_pain.head()
```

```python
chest_pain.plot.bar(x="cp",y="age");
```

```python
chest_pain.plot.bar(x="cp");
```

# Mathplotlib.pyplot

```python
fig, ax=plt.subplots()
ax.bar(chest_pain["cp"],chest_pain["age"],label="age")
ax.set_xlabel("cp")
ax.legend();
```

```python
fig, ax=plt.subplots()
ax.bar(chest_pain["cp"],chest_pain["age"],label="age",edgecolor="k",color=["red","blue","green","black"],)
ax.set_xlabel("cp")
ax.legend();
```

# seaborn

```python

ax=sb.barplot(x="cp",y="age",errcolor="black",capsize=0.1,data=df)
```

# Scatter Plot

Pandas

```python
df.plot.scatter("age","trestbps");
```

# Mathplotlib.pyplot

```python
fig, ax=plt.subplots()
ax.scatter(df["age"],df["trestbps"],marker="^",s= 60,c=df["age"],edgecolor="k",alpha=0.5)
ax.set_xlabel("age")
ax.set_ylabel("trestbps");
```

# Seaborn

```python
sb.scatterplot(x="age",y="trestbps",hue="age",s=30,edgecolor="none",data=df);
```

# Line Plots

# Pandas

```python
ages=df.groupby("age").median().reset_index()
ages.head()
```

```python
ages.plot.line("age","chol");
```

```python
ages.plot.line("age",["chol","trestbps"]);
```
